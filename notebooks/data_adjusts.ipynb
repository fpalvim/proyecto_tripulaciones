```python
# --- Database Connection Settings --
RENDER_DB_URI = os.getenv("RENDER_DB_URI")

# Nombre de la tabla
TABLE_NAME = 'all_events'

# Columna que identifica de forma única cada evento
UNIQUE_ID_COLUMN = 'event_id'

# --- Assuming your DataFrame with the updated data is named 'df' ---

# 1. Filter the DataFrame for the rows you want to send
cinemas_data_to_db = df[df['data_source'] == 'cinemas_list'].copy()

# 2. Establish a connection to your PostgreSQL database using SQLAlchemy
try:
    engine = create_engine(RENDER_DB_URI)
    conn = engine.connect()

    # 3. Iterate through the filtered DataFrame and insert/update data
    for index, row in cinemas_data_to_db.iterrows():
        sql = f"""
        INSERT INTO {TABLE_NAME} (data_source, name, url, tickets_url, ...)
        VALUES (%s, %s, %s, %s, ...)
        ON CONFLICT ({UNIQUE_ID_COLUMN}) DO UPDATE SET
            data_source = EXCLUDED.data_source,
            name = EXCLUDED.name,
            url = EXCLUDED.url,
            tickets_url = EXCLUDED.tickets_url
            -- Add other columns you want to update
        """
        values = (
            row['url'],
            row['tickets_url'],
            row['image']
            # Add values for other relevant columns in the correct order
        )

        try:
            conn.execute(sql, values)
        except Exception as e:
            print(f"Error inserting/updating row {index}: {e}")
            conn.rollback()

    # 4. Commit the changes
    conn.commit()
    print(f"{len(cinemas_data_to_db)} rows from 'cinemas_list' inserted/updated successfully into '{TABLE_NAME}'.")

except Exception as e:
    print(f"Database connection or operation error: {e}")

finally:
    # 5. Close the database connection
    if 'conn' in locals() and conn.is_connected:
        conn.close()
    if 'engine' in locals() and engine:
        engine.dispose()
```
